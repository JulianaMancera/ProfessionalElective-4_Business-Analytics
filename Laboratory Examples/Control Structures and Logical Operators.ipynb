# Control Structures Part 1


## if-else statement

```python
# Test the equality operator

number1 = 7 # Assigning direct value to number1 value
number2 = 8 # Assigning direct value to number2 value

if number1 < number2 :
    print(f"{number1} is less than {number2}")
else:
    print(f"{number1} is greater than {number2}")
```

## if-elif-else

```python
number1 +=5 # add 5 to 7
number2 +=4 # deduct 3 from 8

if number1 < number2:
    print(f"{number1} is less than {number2}")
elif number1 > number2:
    print(f"{number1} is greater than {number2}")
else:
    print(f"{number1} is equal {number2}")
```

```python
# Create a program that will input year level and notify distrinction of year
# Display "freshmen" for number code 1; "sophomore" for code 2; "junor" for code 3; "senior" for code 4

yearLevel = int(input("Enter year level [1-4]: "))

if yearLevel == 1:
    print(f"Your are {yearLevel} level")
    print(f"Catergory as Freshmen")
elif yearLevel == 2:
    print(f"Your are {yearLevel} level")
    print(f"Catergory as Sophomore")
elif yearLevel == 3:
    print(f"Your are {yearLevel} level")
    print(f"Catergory as Junior")
elif yearLevel == 4:
    print(f"Your are {yearLevel} level")
    print(f"Catergory as Senior")
else:
    print(f"{yearLevel} is an invalid code")
```

```python
# Write a python program that determines the age group of a person. A person is considered a 'Child' if his age is between 0-17; 'Young Adult' between age 18-30;
# 'Adult' between 31-59; 'Senior Citizen' between 60 to 100. 'Invalid' ages are negative numbers and more than 100.

age = int(input("Enter age: "))
message = "Invalid age"

if age > 0 and age <= 17:
  message = f"{age} is a child age"
elif age > 17 and age <= 30:
  message = f"{age} is a young adult"
elif age > 30 and age <= 59:
  message = f"{age} is an adult"
elif age > 59 and age <= 100:
  message = f"{age} is a senoir citizen"

print(message)
```

# Control Structure Part 2

## While Loop

```python
# Generate numbers 1 to n. Print only the even number

n = int(input("Enter n: "))
num = 1
while num <= n:
  if not(num % 2):
    print(f"{num}")
  num = num + 1
```

## For Loop


```python
# Generate numbers 1 to n. Print only the odd number. Store numbers to a list. Use for loop

n = int(input("Enter n: "))

list_num = []

for i in range(1, n+1):
  if i % 2:
    list_num.append(i)

print(f"List contains: {list_num}")
```
