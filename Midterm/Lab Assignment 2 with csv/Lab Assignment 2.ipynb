## Date/Time Conversion

1.  Read the patienthistory.csv and store it at df_patient dataframe.
2.  Convert DateOfVisit and LastPaymentDate to datetime objects.
3.  Convert BillAmount and FollowUpDays to float type.
4.  Convert Doctor empty strings into np.nan. (np is a numpy alias)
5.  Display the df_patient
6.  Show df_patient info to check the conversion
7.  Show the total missing values.

```python
# Use pandas library
import pandas as pd

# 1. Read the patienthistory.csv and store it in df_patient dataframe
df = pd.read_csv('PatientHistory.csv')
df
```

```python
# Convert DateOfVisit and LastPaymentDate to datetime objects
df['DateOfVisit'] = pd.to_datetime(df['DateOfVisit'])
df['LastPaymentDate'] = pd.to_datetime(df['LastPaymentDate'])
```

```python
# import the numpy library
import numpy as np
```

```python
# Convert BillAmount and FollowUpDays to float type
df['BillAmount'] = df['BillAmount'].astype(float)
df['FollowUpDays'] = df['FollowUpDays'].astype(float)
```

```python
# Convert Doctor empty strings into np.nan
df['Doctor'] = df['Doctor'].replace('', np.nan)
df
```

```python
# Show df_patient info to check the conversion
df.info()
```

```python
# Show the total missing values.
df.isnull().sum()
```

## Handle Duplicates

8. Check first the total number of duplicates
9. Drop duplicate rows.
10. Store to df_patient_no_duplicates.
11. Print df_patient_no_duplicates.
12.  Check number of missing values in df_patient_no_duplicates

```python
# Check first the total number of duplicates
duplicate_rows = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_rows}")
```

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

```python
# Store to df_patient_no_duplicates
df_patient_no_duplicates = df.copy()
```

```python
# Print df_patient_no_duplicates
df_patient_no_duplicates
```

```python
# Check the number of missing values in df_patient_no_duplicates
df_patient_no_duplicates.isnull().sum()
```

## Median Imputation

13.  Copy df_patient_no_duplicates to df_median
14.  Calculate median of BillAmount stored at median_bill_amount
15. Fill the missing BillAmount with median_bill_amount in df_median
16. Print df_median

```python
# Copy df_patient_no_duplicates to df_median
df_median = df_patient_no_duplicates.copy()
```

```python
# Calculate median of BillAmount stored at median_bill_amount
median_bill_amount = df_median['BillAmount'].median()
print(f"Median of BillAmount: {median_bill_amount}")
```

```python
# Fill the missing BillAmount with median_bill_amount in df_median
df_median['BillAmount'].fillna(median_bill_amount, inplace=True)

# Print df_median
df_median
```

## Mode Imputation

17.  Copy df_median to df_mode
18.  Calculate mode of 'Doctor' and store it at mode_doctor.
19.  Fill missing Doctor with mode_doctor
20.  Print df_mode

```python
# Copy df_median to df_mode
df_mode = df_median.copy()
```

```python
# Calculate mode of 'Doctor' and store it at mode_doctor
mode_doctor = df_mode['Doctor'].mode()[0]
print(f"Mode of Doctor: {mode_doctor}")
```

```python
# Fill missing Doctor with mode_doctor
df_mode['Doctor'].fillna(mode_doctor, inplace=True)

# Print df_mode
df_mode
```

## K-Nearest Neighbor

21. Copy columns 'BillAmount', 'FollowUpDays' stored at columns_for_knn.
22. Copy df_mode with column_for_knn to df_knn
23. Import KNNImputer from sklearn.impute
24. Import numpy with alias np
25. Initialize KNNImputer with imputer variable. Use n_neighbors=5
26. Store to imputed_data after the fit and transform of df_knn
27. Store in df_knn_imputed the array imputted_data after conversion to dataframe. Make sure you retain the original index to map back correctly.
28.  Display df_knn_imputed (FollowUpDays must not contain NaN anymore).  Full dataset must be displayed not just 2 columns.
29.  Print missing values for df_knn_imputed

```python
# Copy columns 'BillAmount', 'FollowUpDays' stored at columns_for_knn
columns_for_knn = ['BillAmount', 'FollowUpDays']
```

```python
# Copy df_mode with column_for_knn to df_knn
df_knn = df_mode[columns_for_knn].copy()
```

```python
# Import KNNImputer from sklearn.impute
from sklearn.impute import KNNImputer
```

```python
# Import numpy with alias np
import numpy as np
```

```python
# Initialize KNNImputer with imputer variable. Use n_neighbors=5
imputer = KNNImputer(n_neighbors=5)
```

```python
# Store to imputed_data after the fit and transform of df_knn
imputed_data = imputer.fit_transform(df_knn)
```

```python
# Store in df_knn_imputed the array imputted_data after conversion to dataframe. Make sure you retain the original index to map back correctly.
df_knn_imputed = pd.DataFrame(imputed_data, columns=columns_for_knn, index=df_knn.index)
```

```python
# Display df_knn_imputed (FollowUpDays must not contain NaN anymore). Full dataset must be displayed not just 2 columns.
# Merge the imputed columns back into the original dataframe
df_imputed_full = df_mode.copy()
df_imputed_full[['BillAmount', 'FollowUpDays']] = df_knn_imputed[['BillAmount', 'FollowUpDays']]

# Display the full dataset
df_imputed_full
```

```python
# Print missing values for df_knn_imputed
df_knn_imputed.isnull().sum()
```
