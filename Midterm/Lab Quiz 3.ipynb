```python
# Import libraries matplotlib, seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Execute the titanic.info to check data types
titanic.info()
```

```python
# Check if you have null values
titanic.isnull().sum()
```

```python
# Describe numerical data
titanic.describe()
```

```python
# Drop duplicates
titanic.drop_duplicates(inplace=True)
```

### 1. What is the distribution of passengers by class?

```python
# Create a countplot to visualize the distribution of passengers by class
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=titanic)
plt.title('Distribution of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

### 2. What is the age distribution of passengers?

```python
# Visualize using histogram to show the distribution of passenger ages
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### 3. How does age vary across passenger classes?

```python
# Use boxplot to visualize the age very across different passenger classes
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Age Distribution Across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

### 4. What percentage of passengers survived, by class or gender?

```python
# Visualize it with the barplot to show the passengers survived bu class or gender
plt.figure(figsize=(10, 6))
sns.barplot(x='class', y='survived', hue='sex', data=titanic, errorbar=None)
plt.title('Survival Rate by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### 5. What is the survival distribution across classes, separated by gender?

```python
# Visualize survival distribution across classes, separated by gender using barplot
plt.figure(figsize=(10, 6))
sns.catplot(x='class', y='survived', hue='sex', kind='bar', data=titanic, errorbar=None)
plt.title('Survival Rate Across Classes by Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### 6. Who survived, grouped by gender?

```python
#  countplot to visualize the number of survived and not survived passengers, separated by gender.
plt.figure(figsize=(8, 6))
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title('Survival Count by Gender')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()
```

### 7. Whatâ€™s the fare distribution for different classes, split by survival?


```python
# Use boxplot to show the fair distribution by different classes but split it by survival
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', hue='survived', data=titanic)
plt.title('Fare Distribution by Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

### 8. How does family size (SibSp -Siblings/Spouse or Parch - Parent/Children) differ across survival outcomes, by gender?


```python
# Family size (SibSp + Parch) difference across survival outcomes, by gender using barplot

titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

plt.figure(figsize=(10, 6))
sns.barplot(x='survived', y='family_size', hue='sex', data=titanic)
plt.title('Average Family Size by Survival and Gender')
plt.xlabel('Survived')
plt.ylabel('Average Family Size')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()
```
